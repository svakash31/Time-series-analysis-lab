```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Step 2: Load the dataset
df = pd.read_csv('/content/Electric_Production.csv')
df.columns = ['Date', 'Production']
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Step 3: Resample monthly (optional, in case data is weekly or daily)
# If data is already monthly, skip this step
# df = df.resample('M').mean()

# Step 4: Apply seasonal decomposition
result = seasonal_decompose(df['Production'], model='additive', period=12)  # monthly data => period=12

# Step 5: Plot the decomposition
plt.figure(figsize=(14, 10))
result.plot()
plt.suptitle('Seasonal Decomposition of Electric Production', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Electric_Production.csv")
df["DATE"] = pd.to_datetime(df["DATE"])
df.set_index("DATE", inplace=True)

df_monthly = df.resample("M").mean()

df["SMA_12"] = df["IPG2211A2N"].rolling(window=12).mean()

plt.figure(figsize=(10,5))
plt.plot(df["IPG2211A2N"], label="Original Data", color="blue", alpha=0.5)
plt.plot(df_monthly, label="Monthly Aggregated", color="red")
plt.plot(df["SMA_12"], label="12-Month Moving Average", color="green")
plt.legend()
plt.title("Trend Estimation & Elimination in Time Series")
plt.xlabel("Year")
plt.ylabel("Production Value")
plt.show()
```
